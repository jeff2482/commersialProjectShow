```python
import pandas as pd
from sklearn.datasets import load_iris
iris = load_iris()
# pandas 兩種型態: DataFrame(2D) Series(1D)
datas = pd.DataFrame(iris["data"], columns = iris["feature_names"])
datas["ans"] = iris["target"]
datas
```

# 資料多寡
看問題難度決定(X欄位的多寡/Y的種類)
1. 機器學習:幾千筆~幾萬筆
2. 深度學習:幾萬筆~幾十萬筆

```python
# pandas 操作:drop(注意axis 0 or 1 橫或直)
# test = pd.DataFrame([[0,1],[2,3]],columns = ["a","b"], index=['a','b'])
# test.drop('a',axis = 1)
x = datas.drop("ans", axis=1)
y= datas["ans"]
y
```

```python
from sklearn.model_selection import train_test_split
# ctrl = "/" :多行註解
# Demo
# train_test_split([1, 2, 3, 4], test_size = 0.25)

#  90% X, 10% x, 90% y, 10% y
x_train,   x_test, y_train, y_test = train_test_split(x, y, test_size=0.1)
print(x_train, x_test, y_train, y_test)
```
